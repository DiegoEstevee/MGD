```python
from TradingviewData import TradingViewData,Interval

request = TradingViewData()
```

```python
request.search('METAL','MCX')
```

```python
nifty_data = request.get_hist(symbol='LINKUSD',exchange='BINANCE',interval=Interval.daily,n_bars=1500)

nifty_data.to_csv("data_csv")
```

```python
import websocket
import json

def on_message(ws, message):
    data = json.loads(message)
    print(data)

ws = websocket.WebSocketApp(
    "https://www.tradingview.com/markets/cryptocurrencies/",
    on_message=on_message
)

ws.run_forever()
```
